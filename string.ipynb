**Exercise4**

How many time 'filoger' is repeated in the following text?

use f-string to print the output!
Uppercase and lowercase letters do not matter.


**'about FILOGER: Filoger is an educational institute that focuses on AI. filoger is the biggest farsispeakers AI community.'**

image.png


```python
i=0
text='about FILOGER: Filoger is an educational institute that focuses on AI. filoger is the biggest farsispeakers AI community.'
text_serche='filoger'

text = text.replace(':','')

text_lower=text.lower()

txt_low_split=text_lower.split()

for str_ in txt_low_split:
  if str_==text_serche:
    i=i+1

f'Filoger is repeated {i} times in the text!'
```
