Exercise 01

Write a program in Python to display the Factorial of a number.

```python
number=int(input('input:'))
fact=number
for i in range(1,number):
    fact=fact*i
print('input: {0}, output: Factorial of {1} is {2}' .format(number,number,fact))
```

**Exercise 02**

Generate random integer number and Write a program to guess the number

```python
import numpy as np
guessed_number=np.random.randint(1,101)
number_input=0
while guessed_number!= number_input:
  number_input=int(input('Generated Number:'))
  if guessed_number<number_input:
    print(' Guess the lower number!')
  elif guessed_number>number_input:
    print('Wrong, Guess the higher number!')

if guessed_number==number_input:
    print('Correct, good job')

```

**Exercise 04**

Use for loop to iterate from 0 to n (input number) and print the sum of all numbers

```python
sum=0
number_input=int(input('input Number:'))
for i in range(1,number_input+1):
  sum=i+sum
print('Input Number:{} ,output: The sum of all numbers is:{}'.format(number_input,sum))
```

**Exercise 05**

This is a fruit list, ['banana','orange','mango','lemon'] reverse the order using loop.


```python
list=['banana','orange','mango','lemon']
for i in range(len(list)):
  last_item=list.pop()
  list.insert(i,last_item)
print(list)
```

**Exercise 06**

Find the max element of the [ 8, 120, 83, 89, 24, 97.5, 201 ]


```python
list=[8, 120, 83, 89, 24, 97.5, 201 ]
max=0
for i in range(len(list)):
    if max<list[i]:
      max=list[i]
      
print('output max ={}'.format(max))
```
